```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
df = pd.read_csv("data/london_companies.csv")

df
```

```python
london_lat = 51.4991568
london_lon = -0.1908626

lon_map = Map(location =[london_lat, london_lon], zoom_start = 11)

for index, row in df.iterrows():
    
    lat_company = row["latitude"]
    lon_company = row["longitude"]
    name_company = row["name"]

    icon = Icon(
                color = "green",
                icon_color = "white",
                icon = "plane",
                prefix="fa-solid fa-building"
        )
    
    marker = Marker(location = [lat_company, lon_company], icon = icon, tooltip=name_company)
    marker.add_to(lon_map)
    
lon_map
```

```python

london_lat = 51.4991568
london_lon = -0.1908626

lon_map = Map(location =[london_lat, london_lon], zoom_start = 11)

locations = []

for index, row in df.iterrows():
    
    lat_company = row["latitude"]
    lon_company = row["longitude"]
    name_company = row["name"]
    
    locations.append([lat_company, lon_company])

    marker = Marker(location = [lat_company, lon_company], tooltip=name_company)
    marker.add_to(lon_map)


heat_map = HeatMap(locations)
heat_map.add_to(lon_map)

lon_map
```
